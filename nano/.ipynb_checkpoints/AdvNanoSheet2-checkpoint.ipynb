```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lam, real, imag = np.loadtxt('Dielectric function of gold.dat.txt', skiprows=1, unpack=True)
```

```python
plt.plot(lam*1e9, real, label='Real part')
plt.plot(lam*1e9, imag, label='Imaginary part')
plt.title('Dielectric Function of Gold')
plt.xlim((200, 2000))
plt.xlabel('Wavelength in nm')
plt.legend(loc='best', title='Legend')
```

```python
plt.plot(lam*1e9, 1/(2*imag))
plt.xlim((200, 2000))
plt.title('Intensity decay length at gold-air interface')
plt.xlabel('Wavelength in nm')
```
