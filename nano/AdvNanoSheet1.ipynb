Aufgabe 3: Plotten von I(x) vs x

```python
p = 1000e-9
w = 500e-9
N = [1,3,5,6,9,11, 51]
```

```python
def a(j):
    return (np.sin(j*np.pi*(w/p)))/(np.pi*j)
```

```python
def I(x, N):
    return  (w/p + sum((a(j)*np.cos(2*np.pi*j*x/p)) for j in range(1, N+1)))**2
```

```python
xAchse = np.linspace(-500e-9, 500e-9, 1000)
```

```python
#plt.yscale('log')
for n in N:
    plt.plot(xAchse, I(xAchse, n), label=str(n))
plt.legend(loc='best', title='Diffraction\nOrders n', fontsize=8)
plt.xlim((-4e-7, 4e-7))
plt.title('Plot of Aerial Image: Intensity across a space')
```
